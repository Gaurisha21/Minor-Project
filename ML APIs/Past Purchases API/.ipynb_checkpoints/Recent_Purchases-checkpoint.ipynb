## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
from mlxtend.frequent_patterns import association_rules
import collections
```

## The Dataset contains the purchase history of a user 

```python
purchases = pd.read_csv('Book1.csv', header = None)
purchases.head()
```

### Converting to list

```python
purchases.values.tolist()
```

### Removing the 'nan'

```python
purchaseList = []
for i in range(len(purchases)):
    purchaseList.append([str(purchases.values[i,j]) for j in range(0, 20) if str(purchases.values[i,j])!='nan'])
purchaseList
```

## Apriori Algorithm for Market Basket Analysis

```python
rules = apriori(purchaseList, min_support = 0.001, min_confidence = 0.03, min_lift = 3, min_length = 1)
```

```python
# a = 'spaghetti'
b = 'french fries'
sugg = []
final = []
i = 0
for item in rules:
    i+=1
    pair = item[0] 
    items = [x for x in pair]
#     print(items)
    if a in items:
#         print(items)
        sugg = [x for x in items]
        final+=sugg
#         print(final)
final = list(set(final))
print("hi ",final)
```

```python
j =0 
data = []
for item in rules:
    print(type(item))
    # First index of the inner list contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    to_print = "Rule: "
    rule = ""

    arrow = " -> "
    for i in range(len(items)):
        to_print += str(items[i]) + arrow
        rule += str(items[i]) + ", "

    print(to_print)
    # Print the support for this association rule
    print("Support: " + str(item[1]))
    # Print the confidence and lift for this association rule
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    data.append([rule,item[1],item[2][0][2],item[2][0][3]])
    print("=====================================")
    j+=1
print("\n\n", j)
print(data)
```

## Most Repeated Purchases with count

```python
SCL = pd.DataFrame(data, columns=['Accociation', 'Support', 'Confidence', 'Lift'])
SCL.head(15)
```
